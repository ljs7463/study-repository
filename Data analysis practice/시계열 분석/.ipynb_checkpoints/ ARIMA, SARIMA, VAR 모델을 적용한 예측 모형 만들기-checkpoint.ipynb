# Data Preprocessing

```python
df=pd.read_csv('https://raw.githubusercontent.com/sm-joo/sm-joo/master/DEXKOUS.csv', parse_dates=['DATE'], index_col='DATE')
```

```python
df.plot(figsize = (12, 5))
```

```python
df.info()
```

2019년도만 살펴보자

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 시각화
# subplot생성
fig, ax = plt.subplots(1,2 , figsize = (12,4))

# acf plot
plot_acf(df, ax = ax[0])

# pacf plot
plot_pacf(df, ax= ax[1])
plt.show()
```

# 파라미터 추정(ARIMA(p,k,q) => p, q 결정)

ARIMA 모델의 잘 알려진 일부 예
- ARIMA(0,0,0)는 백색잡음 모델입니다.
- ARIMA(0,1,0)은 확률보행모델이며, 0이아닌 상수로 구성된 ARIMA(0,1,0)은 표류가 있는 확률보행입니다.
- ARIMA(0,1,1,)는 지수평활 모델이고, ARIMA(0,2,2)는 지수평활추세를 가진 데이터로 확장시킨 홀트의 선형 방법과 같다.(추세를 기반으로 한 예측에 사용될 수 있다.)
- ARIMA(p,0,0) 은 자기회귀 모델, ARIMA(0,0,q)는 이동평균이다.

< 백색잡음 모델 >
- 자기상관(autocorrelation)이 없는 시계열을 백색 잡음(white noise)라고 부른다.
- 자기상관함수(ACF)도표를 통해서 확인해 보면 모두 상관관계가 없는것을 확인해 볼수 있다.
- 시계열 데이터 통계분석에서 백색잡음과정(white noise process) 는 평균이0, 분산이 유한한 상수(sigma^2)인 확률분포로 부터 서로 상관되지 않게 무작위로 샘플을 추출한 신호를 말한다.
- 강한 정상성을 가지고 있는데이터 이다. 

< 확률보행(Random Walk) >
- 확률보행과정은 백색잡음과 달리 '비정상성'을 가지고 있는 데이터이다. 따라서 ARIMA모델에 적용시키기 위해서는 정상화 시켜야하는데 확률보행은 상수가 있는것과 없는것으로 나누어져 있다. 
- 확률보행의 경우 임의의 방향으로 향하는 연속적인 걸음을 나타내는 의미로, 예측 불가능한 변동이 발생하는 것을 뜻한다. 예를들어 위로갈 확률이 0.3 아래로갈 확률이 0.7일경우 주식차트처럼 오르락 내리락 할것이다.




## Graphical method를 통한 시범적인 파라미터 추정

```python
img2 = plt.imread('./이미지 파일/Graphical_method.png')
plt.figure(figsize =(18,8))
plt.imshow(img2)
```

위의 과정에서 차분이전에 ACF, PACF 도표를 다시 돌아가보면 AR의 경우1시차 이후 0으로 절단이 되며, ACF의 경우 지수적으로 감소했고(AR모델로는 stationary), 차분후 MA를 살펴보았을때는 지수적으로 감소하고나 특정시차이후 절단이 없었기때문에 AR(1)모델을 베이스 모델로 사용해 보자(차분1회)

참고로 ARIMA모델을 사용할때 파라미터를 선정하는데 있어서 p*q = 0이고 p+q < 2 인 값을 많이쓴다고 한다. 

그리고 시계열 데이터가 AR특성을 띄는 경우 위처럼 ACF는 서서히 감소하고, PACF는 p시차를 제외하고 급격히 감소하며, MA의 경우 ACF는 q시차 이후 급격히 감소하고 PACF는 천천히 감소한다. 급격히 감소하는 시차를 각각  AR과 MA의 모수(p,q)로 사용가능하다.

## ARIMA 임시 모델 적용하기


파라미터 설명
- order : AR차수, 차분횟수, MA차수
- freq : 날짜 

    'B' - business day, ie., Mon. - Fri.
    'D' - daily
    'W' - weekly
    'M' - monthly
    'A' - annual
    'Q' - quarterly
- trend : 상수항 포함여부, 'c'(constant) or 'nc'(no constant) -> 'c'를 했을때 constant가 0.05보다 크게나온다면 nc를 하는게 좋다. 

```python
########
# 여기서 안정성 있으려면 1회차분이 있어야 했었는데, AR(1)모델의 경우 차분하기전에 AR(1)이 적합으로 나왔는데 그렇다면 차분을 넣어야
# 하는가 빼야하는가
```

```python
model =ARIMA(train_df, order = (1,1,0), freq = 'B')
model_fit = model.fit()
model_fit.summary()
```

AR(1)모델을 돌려본 결과 AR(1)은 p-value가 0.05보다 커서 유의하지 않는 파라미터 값으로 나오게되었습니다.

## 더 좋은 파라미터 찾기

```python
# Parameter search

p = range(0,3)
d = range(0,3)
q = range(0,3)
pdq = list(itertools.product(p, d,q))

aic = []
for i in pdq:
    model = ARIMA(train_df, order = (i))
    model_fit = model.fit()
    print(f'ARIMA: {i} >> AIC : {round(model_fit.aic,2)}')
    aic.append(round(model_fit.aic,2))
```

```python
# Search optimal parameters
optimal = [(pdq[i],j) for i, j in enumerate(aic) if j == min(aic)]
optimal
```

```python
model =ARIMA(train_df, order = optimal[0][0], freq = 'B')
model_fit = model.fit()
model_fit.summary()
```

# 모델 적합도 체크(Diagnosis check)

## 잔차들의 ACF도표를 통한 확인

```python
plot_acf(model_fit.resid)
plt.show()
```

## Argumentd dickey fuller test

```python
adfuller(model_fit.resid)
```

p-value가 0.05보다 작은것으로 보아 귀무가설인 non-staionary한 시계열인것을 기각하고 정상 시계열임을 채택할 수 있다.

# 예측해보기

arima계열 데이터는 데이터가 너무많으면 예측하기에 별로안좋다??

```python
model = ARIMA(train_df, order = (2,1,0), freq = 'B')
model_fit = model.fit()
```

```python
result = model_fit.get_forecast(steps = len(test_df.index), alpha = 0.05).summary_frame()
result
```

```python
fc = tuple(result['mean'].values) # 예측값
se = tuple(result['mean_se'].values) # 표준오차
lower_coef = tuple(result['mean_ci_lower'].values) # 신뢰구간 최소
upper_coef = tuple(result['mean_ci_upper'].values) # 신뢰구간 최대

# 예측값 인덱스 넣기
fc_data = pd.Series(fc, index = test_df.index)

# 신뢰구간 인덱스 넣기
lower_data = pd.Series(lower_coef, index = test_df.index)
upper_data = pd.Series(upper_coef, index = test_df.index )
```

```python
# VIZ
plt.figure(figsize =(15, 6))
plt.plot(train_df, label = 'training')
plt.plot(test_df, label = 'actual')
plt.plot(fc_data, label = 'forecast')
plt.fill_between(test_df.index, lower_data, upper_data, color = 'black', alpha = 0.1)
plt.legend(loc = 'upper left')
plt.show()
```
