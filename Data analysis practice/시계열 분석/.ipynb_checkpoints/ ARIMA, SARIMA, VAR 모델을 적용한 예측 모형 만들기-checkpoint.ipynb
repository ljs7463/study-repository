# 파라미터 추정(ARIMA(p,k,q) => p, q 결정)

ARIMA 모델의 잘 알려진 일부 예
- ARIMA(0,0,0)는 백색잡음 모델입니다.
- ARIMA(0,1,0)은 확률보행모델이며, 0이아닌 상수로 구성된 ARIMA(0,1,0)은 표류가 있는 확률보행입니다.
- ARIMA(0,1,1,)는 지수평활 모델이고, ARIMA(0,2,2)는 지수평활추세를 가진 데이터로 확장시킨 홀트의 선형 방법과 같다.(추세를 기반으로 한 예측에 사용될 수 있다.)
- ARIMA(p,0,0) 은 자기회귀 모델, ARIMA(0,0,q)는 이동평균이다.

< 백색잡음 모델 >
- 자기상관(autocorrelation)이 없는 시계열을 백색 잡음(white noise)라고 부른다.
- 자기상관함수(ACF)도표를 통해서 확인해 보면 모두 상관관계가 없는것을 확인해 볼수 있다.
- 시계열 데이터 통계분석에서 백색잡음과정(white noise process) 는 평균이0, 분산이 유한한 상수(sigma^2)인 확률분포로 부터 서로 상관되지 않게 무작위로 샘플을 추출한 신호를 말한다.
- 강한 정상성을 가지고 있는데이터 이다. 

< 확률보행(Random Walk) >
- 확률보행과정은 백색잡음과 달리 '비정상성'을 가지고 있는 데이터이다. 따라서 ARIMA모델에 적용시키기 위해서는 정상화 시켜야하는데 확률보행은 상수가 있는것과 없는것으로 나누어져 있다. 
- 확률보행의 경우 임의의 방향으로 향하는 연속적인 걸음을 나타내는 의미로, 예측 불가능한 변동이 발생하는 것을 뜻한다. 예를들어 위로갈 확률이 0.3 아래로갈 확률이 0.7일경우 주식차트처럼 오르락 내리락 할것이다.




## Graphical method를 통한 시범적인 파라미터 추정

```python
img2 = plt.imread('./이미지 파일/Graphical_method.png')
plt.figure(figsize =(18,8))
plt.imshow(img2)
```

위의 과정에서 차분이전에 ACF, PACF 도표를 다시 돌아가보면 AR의 경우1시차 이후 0으로 절단이 되며, ACF의 경우 지수적으로 감소했고(AR모델로는 stationary), 차분후 MA를 살펴보았을때는 지수적으로 감소하고나 특정시차이후 절단이 없었기때문에 AR(1)모델을 베이스 모델로 사용해 보자(차분1회)

참고로 ARIMA모델을 사용할때 파라미터를 선정하는데 있어서 p*q = 0이고 p+q < 2 인 값을 많이쓴다고 한다. 

그리고 시계열 데이터가 AR특성을 띄는 경우 위처럼 ACF는 서서히 감소하고, PACF는 p시차를 제외하고 급격히 감소하며, MA의 경우 ACF는 q시차 이후 급격히 감소하고 PACF는 천천히 감소한다. 급격히 감소하는 시차를 각각  AR과 MA의 모수(p,q)로 사용가능하다.

## 더 좋은 파라미터 찾기

##### 추후에 해결할점 : AIC와 likelihood의 값이 각각 최소, 최고가 되는 파라미터를 정하게되었을 때 p-vlaue가 좋지 않은 경우 이 모델에 대해 어떻게 평가해야하는가..

```python
p = range(0, 3)
d = range(0, 2)
q = range(0, 3)
pdq = list(itertools.product(p, d, q))

dict_model = {}
for i in pdq:
    
    try:
        model = ARIMA(train_df, order = (i))
        print(i)
        model_fit = model.fit()
        dict_model[i] = [model_fit.llf, model_fit.aic]
    except:
        print('error lag:', i)
        
```

```python
params_df = pd.DataFrame.from_dict(dict_model, orient = 'index',columns = ['llf','aic'])
params_df
```

```python
params_df[params_df['llf']==params_df['llf'].max()]
```

```python
params_df[params_df['aic']==params_df['aic'].min()]
```

likelihood 로 보았을때는 ARIMA(2,0,2)모델이 , AIC 로 보았을땐 ARIMA(2, 0, 0) 가장좋은 모델이다. LLR TEST를 통해서 Likelihood가 유의미 하게 다른지 검증해보고 모델을 선택해보자.

```python
from scipy.stats.distributions import chi2
L1 = 1261.013945
L2 = 1259.366313
LR = (2*(L2-L1))
p = chi2.sf(LR, 2)
p
```

LLR테스트 결과 두개의 Likelihood는 다를 확률이 높기때문에 각각 적용해서 p-value도 확인해보자

```python
tmp_model1 =ARIMA(train_df, order =(2,1,2), freq = 'B')
model_fit1 = tmp_model1.fit()
model_fit1.summary()
```

```python
tmp_model2 =ARIMA(train_df, order =(0,1,2), freq = 'B')
model_fit2 = tmp_model2.fit()
model_fit2.summary()
```

파라미터의 p-value는 ARIMA(0,1,2)모델이 조금 더 좋게 나온다는것을 알 수 있다. 이때는 aic를 선택하는게 맞을까??

# 예측해보기

arima계열 데이터는 데이터가 너무많으면 예측하기에 별로안좋다??

```python
model = ARIMA(train_df, order = (2,1,2), freq = 'B')
model_fit = model.fit()
result = model_fit.get_forecast(steps = len(test_df.index), alpha = 0.05).summary_frame()

fc = tuple(result['mean'].values) # 예측값
se = tuple(result['mean_se'].values) # 표준오차
lower_coef = tuple(result['mean_ci_lower'].values) # 신뢰구간 최소
upper_coef = tuple(result['mean_ci_upper'].values) # 신뢰구간 최대

# 예측값 인덱스 넣기
fc_data = pd.Series(fc, index = test_df.index)

# 신뢰구간 인덱스 넣기
lower_data = pd.Series(lower_coef, index = test_df.index)
upper_data = pd.Series(upper_coef, index = test_df.index )

# VIZ
plt.figure(figsize =(15, 6))
plt.plot(train_df, label = 'training')
plt.plot(test_df, label = 'actual')
plt.plot(fc_data, label = 'forecast')
plt.fill_between(test_df.index, lower_data, upper_data, color = 'black', alpha = 0.1)
plt.legend(loc = 'upper left')
plt.show()
```

```python
model = ARIMA(train_df, order = (0,1,2), freq = 'B')
model_fit = model.fit()
result = model_fit.get_forecast(steps = len(test_df.index), alpha = 0.05).summary_frame()

fc = tuple(result['mean'].values) # 예측값
se = tuple(result['mean_se'].values) # 표준오차
lower_coef = tuple(result['mean_ci_lower'].values) # 신뢰구간 최소
upper_coef = tuple(result['mean_ci_upper'].values) # 신뢰구간 최대

# 예측값 인덱스 넣기
fc_data = pd.Series(fc, index = test_df.index)

# 신뢰구간 인덱스 넣기
lower_data = pd.Series(lower_coef, index = test_df.index)
upper_data = pd.Series(upper_coef, index = test_df.index )

# VIZ
plt.figure(figsize =(15, 6))
plt.plot(train_df, label = 'training')
plt.plot(test_df, label = 'actual')
plt.plot(fc_data, label = 'forecast')
plt.fill_between(test_df.index, lower_data, upper_data, color = 'black', alpha = 0.1)
plt.legend(loc = 'upper left')
plt.show()
```

# 그외 모델 알아보기

ARIMAX : 일종의 회귀모형으로 볼 수 있다. 일번적인 AR이나 MA모형의 경우 단변량 시계열을 표현하는데 적절하다. 하지만, ARIMAX모형은 추가적인 외생 변수(Explanatory variable)을 활용함으로서 다변량 시계열 데이터를 활용하기에 적절한 모형디ㅏ.

SARIMA : 말그대로 계절성(Seasonal)을 포함하는 모델이다. SARIMA(p,d,q)(P,D,Q,S)로  표현한다.
    - p : order of non-seasonal AR terms
    - d : order of non-seasonal differencing
    - q : order of non-seasonal MA terms
    - P : order of seasonal AR terms
    - D : order of seasonal differencing
    - Q : order of seasonal MA terms
    - S : length of the cycle
    
SARIMAX : ARIMAX와 마찬가지로 기존 SARIMA 모형에, 외생변수 X를 포함하여 모형을 수립한다. 
    - SARIMA나 SARIMAX의 경우에는 객관적으로 차수를 정하는 것이 어려운 것 같다. 따라서 실무에서 많은 경우 grid search를 
    활용하여  AIC, BIC, 예측력을 기준으로 차수를 정하는 경우도 많다.


## SARIMA 모델

ARIMA 모형은 주기성 및 계절성이 강하거나 불규칙한 데이터에 한계를 보이고 있어 교통량 데이터와 같
은 자료에 적용하기에는 한계가 있다. 이러한 한계점을 극복하고 보완하기 위해 ARIMA 모형에 자기상관 모
형인 SAR(Seasonal Auto Regressive)과 계절 이동평균 모형인 SMA(Seasonal Moving Average)가 결합된 분석
기법이 SARIMA 모형이다.

ARIMA가 추세를 통해서 예측을 하는 모델이며 정상성을 만족해야하기 때문에 non-seasonality 하지만, SARIMA는 Seasonal이 추가가된다 즉, 추세성 + 계절성으로 예측을 하는 모델이다.




```python
# Parameter grid search
p = range(0, 3)
d = range(0, 2)
q = range(0, 3)
pdq  = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2],12) for x in list(itertools.product(p, d, q))]

dict_model = {}

for i in pdq:
    for j in seasonal_pdq:
        try:
            model = SARIMAX(train_df, order = (i), seasonal_order = (j))
            print(i, j)
            model_fit = model.fit()
            dict_model[i] = [model_fit.llf, model_fit.aic]
        
        except:
            continue
            
```

```python
params_df = pd.DataFrame.from_dict(dict_model, orient = 'index',columns = ['llf','aic'])
params_df
```
