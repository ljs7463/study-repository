시계열 특성 > 안정성(Stationary, 정상성)
==> 시계열 데이터가 미래에 똑같은 모양일 확률이 매우 높다.

정상성 : 시간의 흐름에 따라 일정한 분산과 일정한 평균을 가지고 있음, 데이터가 정상성을 가지고 있다는 뜻은 평균과 분산이 안정화되어 분석하기 쉽다라는 뜻이다.


즉, 시계열이 안정적이지 않으면 현재의 패턴이 미래에 똑같이 재현되지 않으므로, 그대로 예측 기법을 즉용하면 안된다.

불안정한 시계열을 그대로 예측하게되면, 설명력(R2)가 90%이상, 정확도 90% 이상이 나온다. 그러나, Spurious regression(가성적회귀귀), Overfitting -> 따라서 안정정인(stationary)하게 만든 후 예측을 해야한다.

<시계열 안정성 판별볍 >
 
 1. "Dickey Fuller Test"

   - 귀무가설 : 원계열은 안정적이지 않다.
  - 테스트 결과 해석 
     - p-value 가 0.05보다 작으면, 귀무가설 기각. 즉, 안정적인 시계열
      - p-value 가 0.05보다 크면, 귀무가설 채택. 즉, 불안정한 시계열

2. ACF plot 과 PACF plot
 



불안정한 시계열을 안정적인 시계열로 변경하는 가장 보편적인 방법은 성장률로 변화해서 예측하는 것입니다.



<요약>
- 시계열 안정성 테스트 -> Augmented Dickey Fuller Test
- 불안정한 데이터를 안정적 데이터로 변환 -> 성장률, log difference

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# plot할때 warning message가 난다면, 아래 명령어로 matplotlib 버전을 3.0.3으로 낮춰주세요.
# colab에 설치된 matplotlib은 2020년4월 기준으로 3.2.1 입니다. 
! pip install matplotlib==3.0.3

# 설치된 matplotlib 버전을 확인하려면 아래를 실행하세요.
import matplotlib
matplotlib.__version__
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/sm-joo/sm-joo/master/DEXKOUS.csv', parse_dates=['DATE'], index_col='DATE')
df.columns=["KOUS"]
df['KOUS'].replace('.', '', inplace=True)
df['KOUS'] = pd.to_numeric(df['KOUS'])
df.fillna(method='ffill', inplace=True)
```

## Resample, Rolling

# 연도별 

```python
import matplotlib.pyplot as plt
df['KOUS'].plot()
```

```python
# resample : 일별데이터 -> 주단위 데이터, 월단위 데이터로 변환
df.resample('M').last()
```

```python
df.resample('M').last().plot()
```

```python
df.resample('W-Fri').last()
```

```python
# rolling : 이전 xx일에 대한 이동평균, 이동sum을 산출할때 사용
df.rolling(100).mean().plot()
```

```python
# 매월 말 기준 앞에있는 30일 기준으로 표준편차
df.rolling(30).std().resample('M').mean().plot() 
```

## 안정성 검정(ADF Test)
- 귀무가설 = 안정적이지 않다.
- p-value가 0.05보다 작으면, 귀무가설 기각. 즉, 안정적인 시계열
- p-value가 0.05보다 크면, 귀무가설 채택. 즉, 불안정한 시계열
- 안정적인 데이터로 변경 : 변화율 / 로그차분

```python
adfuller(df['KOUS'])[1]
```

```python
plot_acf(df)
plt.show()
```

```python
plot_pacf(df)
plt.show()
```

귀무가설 채택 -> 안정적이지 않다

# 인사이트

```python
df.plot(figsize = (16,5))
```

```python
# 주단위로 변경해서 보기
df_w = df.resample('W-Fri').last()
```

```python
# 2017년 2019년
df_2017 = df_w[df_w.index.year==2017]
df_2019 = df_w[df_w.index.year==2019]
```

```python
df_2017.plot()
```

```python
df_2019.plot()
```

```python
plot_acf(df_2017)
plt.show()
```

2017년 하나의 외부요인이 3주정도 지속된다.( 상관관계가 시차가 3인 시점까지 높다는것은 시차가 3인 시점까지 관계가 있다 즉, 하나의 외부요인이 3인 시차까지 영향을 미친다는 뜻이고, 해당데이터에서는 3주를 의미한다.)

```python
plot_acf(df_2019)
plt.show()
```

2019년에는 외부요인이 4주까지 지속된다

```python
# 첫번째 행 : 2017년 데이터의 원계열, ACF, PACF
# 두번째 행 : 2019년 데이터의 원계열, ACF, PACF
fig, ax = plt.subplots(2,3,figsize = (16, 7))
ax[0,0].plot(df_2017)
ax[0,0].set_title('original series(2017')
ax[1,0].plot(df_2019)
ax[1,0].set_title('original series(2017')
plot_acf(df_2017, ax = ax[0,1])
plot_acf(df_2019, ax = ax[1,1])
plot_pacf(df_2017, ax = ax[0,2])
plot_pacf(df_2019, ax = ax[1,2])
plt.show()
```

- 2017년에 비해 2019년은 외부 충격이 오래 지속되었다.
- 2017년에는 외부충격이 다음기에 0.75 남아있지만, 2019년에는 0.9남아있다.
(persistency가 증가하고 있다.)
--> 가입자, 사용자 마케팅효과 분석
--> 주가지수, 환율 : 외부 충격이 얼마나 오래 지속되는가.
