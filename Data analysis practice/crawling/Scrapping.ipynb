## HTML 태그 검색

- find() : 해당 조건에 맞는 하나의 태그를 가져온다.
- find_all() : 해당 조건에 맞는 모든 태그를 가져온다.(리스트 형태로 반환)
- select_one() : CSS선택자와 같은 형식으로 선택 하여 조건에 맞는 하나
- select() : CSS 선택자와 같은 형식으로 선택 가능

#### Find와 Select차이

1. find는 태그 요소를 이용해서 찾는 방법, select는 css를 이용해서 찾는 방법
2. select의 장점 : 하위로 이동이 편리함, 특정 요소로 이동이 편리함

tag = "< p class='practice' id='seok'> 데이터분석가 < / p>" learn = Machinelearning(tag) 

### 1) find

##### 태그 이름만특정
soup.find('p')
##### 태그  속성만 특정
soup.find(class_=''practice')/ soup.find(attrs = {'class':'practice'})
##### 태그 이름과 속성 모두 특정
soup.find('p', class_ = 'practice')

### 2) select

##### 태그 이름만 특정
soup.select_one('p')
##### 태그 class 특정
soup.select_one('.practice')
##### 태그 이름과 class 모두 특정
soup.select_one('p.practice')
##### 태그 id 특정
soup.select_one('#seok')
##### 태그 이름과  id 모두 특정
soup.select_one('p#seok')
##### 태그 이름과 class, id 모두 특정
soup.select_one('p.practice#seok')

## 검색키워드를 입력받아 검색된 페이지의 제목들을 크롤링 해주는 코드

```python
# 검색어를 다양하게 입력해도 원하는 구인공고 보여주는 크롤링(find 사용)
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup

Base_url = 'https://www.saramin.co.kr/zf_user/search?search_area=main&search_done=y&search_optional_item=n&searchType=search&searchword='
Plus_url = input('검색어를 입력하세요: ')
url = Base_url +quote_plus(Plus_url) # 한글을 검색하게 되면  urllib.parse.quote_plus을 앞에 붙여줘야한다.
html = urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
title = soup.find_all(class_ = 'job_tit')

for i in title:
    print(i.a['title'])
    print('https://www.saramin.co.kr' + i.a['href'])
    print()
```

```python
# 검색어를 다양하게 입력해도 원하는 구인공고 보여주는 크롤링(select 사용)
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup

Base_url = 'https://www.saramin.co.kr/zf_user/search?search_area=main&search_done=y&search_optional_item=n&searchType=search&searchword='
Plus_url = input('검색어를 입력하세요: ')
url = Base_url + quote_plus(Plus_url) # 한글을 검색하게 되면  urllib.parse.quote_plus을 앞에 붙여줘야한다.
html =  urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
title = soup.select('.job_tit a')
for i in title:
    print(i['title'])
    print('https://www.saramin.co.kr' + i['href'])
    print()
```

```python
from selenium import webdriver
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
baseUrl = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=image&query='
plusUrl = input('검색어를 입력해 주세요 :')
url =baseUrl + quote_plus(plusUrl)
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

img = soup.find_all('section', class_='sc_new.sp_nimage._prs_img._imageSearchPC')
breakpoint()
for i in img:
    print(img.a.attrs.get('src'))
    print(i.select_one('src').text)
    
driver.close()
```
